# 중위값

```python
num_fin = pd.read_csv("금융기관수.csv",engine="python")
```

```python
num_fin
```

```python
price = pd.read_excel("중위단위매매가격_아파트.xlsx",header=10,encoding="euc-kr")
```

```python
price1=price.loc[:,["구별","2017년 12월"]]
```

```python
price1.drop([0],inplace=True)
```

```python
price1
```

```python
result_data = pd.merge(price1,num_fin,on="구별")
result_data
```

```python
result_data.sort_values(by="금융기관",ascending=False)
```

```python
result_data=result_data.set_index("구별")
```

```python
result_data.describe()
```

# 회귀분석

```python
## g

fp1 = np.polyfit(result_data['금융기관'],result_data['2017년 12월'],1) ## (x축,7축,1차함수)
```

```python
fp1 ## array(기울기, y절편)
```

```python
f1 = np.poly1d(fp1) ## 선형화시키는 함수 poly1d
```

```python
f1
```

```python
np.corrcoef(result_data["금융기관"],result_data["2017년 12월"]) ##  상관관계
```

```python
x = np.linspace(20,350,20)   # 회귀선 X축 설정
```

```python
# 시각화

plt.figure(figsize=(10,10))
plt.scatter(result_data["금융기관"],result_data["2017년 12월"]) ## 점 들표시
plt.plot(x,f1(x),ls="dashed",lw=3,c="r") ## 회귀선 
plt.title("매매가격과 은행점포 수의 관계",size=20)
plt.xlabel("금융기관수",size = 15)
plt.ylabel("평당 매매가격",size =15)
plt.grid()
plt.show()   
```

```python
## 회귀선과 가장 멀리 떨어진 정도 ~ 오차

result_data["오차"] = np.abs(result_data["2017년 12월"]-f1(result_data["금융기관"]))
sort_result = result_data.sort_values(by="오차",ascending=False)
sort_result.head()
```
